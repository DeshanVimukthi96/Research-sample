```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split,cross_val_score,GridSearchCV,KFold
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.impute import KNNImputer
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from sklearn.metrics import confusion_matrix, accuracy_score,classification_report,roc_curve,roc_auc_score
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("C:\\Users\\Dedshan Vimukthi\\Documents\\Dataset2\\ori_ckd.CSV")
```

```python
data.head()
```

```python
data.drop("id",axis=1,inplace=True)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
new_data=pd.get_dummies(data,columns=["gender","class"],drop_first=True)
```

```python
new_data.head()
```

```python
data_num=new_data[["age","weight","cr","cru","al","gl","na","ca","k","cl","pr","iron","ua","wbc","rbc","hg","plt"]]
```

```python
#imputer = KNNImputer(n_neighbors=3)
imputer = KNNImputer(n_neighbors=5)
```

```python
df = pd.DataFrame(imputer.fit_transform(data_num),columns = data_num.columns)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
ssc=StandardScaler()
```

```python
sscnum=ssc.fit_transform(df)
```

```python
sscnum
```

```python
num_data=pd.DataFrame(sscnum,columns=["age","weight","cr","cru","al","gl","na","ca","k","cl","pr","iron","ua","wbc","rbc","hg","plt"])
```

```python
num_data.head()
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(num_data.corr(),annot=True,vmin=-1,vmax=1)
plt.show()
```

```python
num_data.drop("ua",axis=1,inplace=True)
num_data.drop("rbc",axis=1,inplace=True)
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(num_data.corr(),annot=True,vmin=-1,vmax=1)
plt.show()
```

```python
cat_data=new_data[["gender_M","class_1"]]
```

```python
data=pd.concat([num_data, cat_data], axis=1)
data.head()
```

```python
data.corr()
```

```python
data.shape
```

```python
x=data.iloc[:,:18]
y=data.iloc[:,18]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
knncl = KNeighborsClassifier(n_neighbors=5)
knncl.fit(x_train, y_train)
```

```python
y_pred=knncl.predict(x_test)
```

```python
#knn3
accuracy_score(y_test,y_pred)
```

```python
#knn5
accuracy_score(y_test,y_pred)
```

```python
rfcl = RandomForestClassifier()
rfcl.fit(x_train, y_train)
```

```python
y_pred=rfcl.predict(x_test)
```

```python
#knn3
accuracy_score(y_test,y_pred)
```

```python
#knn5
accuracy_score(y_test,y_pred)
```

```python
svc = SVC(C=1, kernel="linear")
```

```python
svc.fit(x_train,y_train)
```

```python
y_pred=svc.predict(x_test)
```

```python
#knn3
accuracy_score(y_pred,y_test)
```

```python
#knn5
accuracy_score(y_pred,y_test)
```

```python
svc = SVC(C=1, kernel="rbf")
```

```python
svc.fit(x_train,y_train)
```

```python
y_pred=svc.predict(x_test)
```

```python
#knn5
accuracy_score(y_pred,y_test)
```
